Ejercicio 1

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Datos
datos = [2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6, 3, 3, 2, 1, 1, 5, 6, 7, 4, 3, 2, 5, 4, 6]

# Intervalos de amplitud 2
bins = np.arange(0, 11, 2)

# Histograma
hist, edges = np.histogram(datos, bins=bins)

# Tabla de frecuencias
frecuencia_df = pd.DataFrame({
    'Intervalos': [f'[{edges[i]}, {edges[i+1]})' for i in range(len(edges)-1)],
    'Frecuencia Absoluta': hist
})

print(frecuencia_df)

# Histograma
plt.hist(datos, bins=bins, edgecolor='black', color='skyblue')
plt.title('Histograma de Frecuencia de Datos')
plt.xlabel('Intervalos')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(edges)
plt.show()
```

Ejercicio 9

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Frecuencia de alumnos por intervalo
intervalos_peso = ['[50, 55)', '[55, 60)', '[60, 65)', '[65, 70)', '[70, 75]']
alumnos = [10, 40, 25, 20, 5]

# Crear los datos de los alumnos repetidos según su frecuencia en los intervalos
datos_pesos = [50] * 10 + [55] * 40 + [60] * 25 + [65] * 20 + [70] * 5

# Crear el histograma con los datos generados
plt.hist(datos_pesos, bins=[50, 55, 60, 65, 70, 75], edgecolor='black', color='lightcoral')

# Etiquetas y título
plt.title('Histograma de Pesos de Alumnos')
plt.xlabel('Peso (kg)')
plt.ylabel('Número de Alumnos')

# Mostrar el gráfico
plt.show()

# Crear la tabla de frecuencias
tabla_frecuencia = pd.DataFrame({
    'Intervalos': intervalos_peso,
    'Frecuencia Absoluta': alumnos
})

print(tabla_frecuencia)
```
